# Code Challenge: Linearity of Trace

### Objective
Verify the linearity of the `trace` operator using two primary axioms:
1. **Additivity:** $Tr(A + B) = Tr(A) + Tr(B)$
2. **Homogeneity:** $Tr(\lambda A) = \lambda \cdot Tr(A)$

### Complexity Analysis
* **Time Complexity:** $O(n)$ — The trace operation only visits the $n$ diagonal elements of an $n \times n$ matrix, ignoring the other $n^2 - n$ elements.
* **Space Complexity:** $O(n^2)$ — Storing the two random input matrices.

### Implementation Notes
* Handled using `numpy.trace()` and verified with `numpy.allclose()` to account for floating-point precision.

```python
# Setup square matrices
m, n = 3, 3
A = np.random.randn(m, n)
B = np.random.randn(m, n)
l = 5

# 1. Test Additivity: Tr(A) + Tr(B) == Tr(A + B)
sum_of_traces = np.trace(A) + np.trace(B)
trace_of_sum   = np.trace(A + B)
print(f"Additivity Test Output: {sum_of_traces, trace_of_sum}")

# 2. Test Homogeneity: Tr(l*A) == l*Tr(A)
trace_of_scalar_mult = np.trace(l * A)
scalar_times_trace   = l * np.trace(A)
print(f"Homogeneity Test Output: {trace_of_scalar_mult, scalar_times_trace}\n")

# Verification Output
print(f"Additivity holds: {np.allclose(sum_of_traces, trace_of_sum)}")
print(f"Homogeneity holds: {np.allclose(trace_of_scalar_mult, scalar_times_trace)}")
```
