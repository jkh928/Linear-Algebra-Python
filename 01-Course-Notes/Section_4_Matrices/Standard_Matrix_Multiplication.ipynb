```python
# ======================================================================
# COMPLEXITY: O(m * n * p) - Dot product of each row A with each column B.
# DATA SHAPE: A(m, n) @ B(n, p) -> Result(m, p). Inner 'n' must match.
# MATH: Non-Commutative -> AB != BA. Associative -> (AB)C = A(BC).
# TOOLS: A @ B, np.matmul(A, B), A.dot(B)
# ======================================================================

import numpy as np

# set global print options
np.set_printoptions(suppress=True, precision=3)
```

### Standard Matrix Multiplication

The Rule: To multiply matrix $A$ by matrix $B$, the number of columns in $A$ must equal the number of rows in $B$.
* Shape Logic: $(m \times n) \times (n \times p) \rightarrow (m \times p)$
* The "Inner" Dimensions: Must match ($n$).
* The "Outer" Dimensions: Define the shape of the resulting matrix ($m \times p$).

🛠️ NumPy Implementation
Standard matrix multiplication (the Dot Product) is performed using @ or np.matmul() or A.dot(B).
* This is not the same as A * B (which is element-wise).

```python
# create matrices

m = 2
n = 5
k = 7

A = np.random.randn(m,n) # 2x5 matrix
B = np.random.randn(n,k) # 5x7 matrix
C = np.random.randn(m,k) # 2x7 matrix

print(f"2x5 * 5x7 = 2x7 Matrix: \n{np.matmul(A,B)}\n")
print(f"5x2 * 2x7 = 5x7 Matrix: \n{A.T@C}\n")
print(f"5x7 * 7x5 = 5x5 Matrix: \n{B.dot(B.T)}\n")
print(f"7x5 * 5x7 = 7x7 Matrix: \n{(B.T@B)}\n")
print(f"2x7 * 7x5 = 2x5 Matrix: \n{(C@B.T)}\n")
```
