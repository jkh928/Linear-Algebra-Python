```python
import numpy as np

# set global print options
np.set_printoptions(suppress=True, precision=3)
```

### 1. Row-wise Broadcasting (The Default)
When a 1D vector's length matches the number of columns in a matrix, NumPy automatically "stretches" that vector down to every row.

Matrix A: (5, 4)

Vector c: (4,)

Result: (5, 4) + (4,) → Success

#### 2. Column-wise Broadcasting (The Reshape)
To add a vector to every column, the vector must match the number of rows and must be explicitly 2D.

Matrix A: (5, 4)

Vector f: (5, 1)

Result: (5, 4) + (5, 1) → Success

```python
# create a matrix
A = np.reshape(np.arange(1, 21), (5, 4), 'F') # F = by column, C = by row (default)

# create vectors
c = [12, 25, 100, 200]
f = [24, 50, 200, 400, 800]

print(f"Original Matrix: \n{A}\n")
print(f"Row Vector: \n{c}\n")
print(f"Column Vector: \n{np.reshape(f, (len(f),1))}\n")

# broadcast on rows
print(f"Broadcast on Rows: \n{A+c}\n")
print(f"Broadcast on Columns: \n{A+np.reshape(f, (len(f), 1))}\n")
```
