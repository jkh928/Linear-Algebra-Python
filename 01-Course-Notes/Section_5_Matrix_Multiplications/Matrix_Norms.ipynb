### Matrix Norms (Common)

#### 📌 Overview
In Linear Algebra, matrix norms provide a way to measure the "size" or "magnitude" of a matrix. Different norms measure different aspects of the matrix, such as its energy, its maximum stretching factor, or the sum of its singular values.

#### 🛠️ Norm Categories
##### 1. Frobenius Norm (fro)
This is the most common norm, equivalent to the Euclidean norm of a vector. It is calculated as the square root of the sum of the absolute squares of its elements.
* Math: $\|A\|_F = \sqrt{\sum_{i,j} |a_{ij}|^2}$
* Utility: Represents the total "energy" of the matrix.

##### 2. Induced 2-Norm (L2)
The induced 2-norm (or spectral norm) represents the maximum amount a matrix can "stretch" a vector. It is the most important norm for analyzing matrix stability and error propagation.
* Math: $\|A\|_2 = \sigma_{max}(A) = \sqrt{\lambda_{max}(A^T A)}$
* Verification: In this notebook, we verify this by comparing np.linalg.norm against the square root of the maximum eigenvalue of the Gram matrix.

##### 3. Schatten p-Norm
The Schatten norm is a generalization that operates on the singular values of a matrix rather than its elements.
* p=1 (Nuclear Norm): The sum of the singular values. This is used frequently in Machine Learning for matrix completion and regularization.
* p=2: Equivalent to the Frobenius norm.

```python
# Create a random matrix with orthogonal columns
matrix_size = 5
q_matrix, r_matrix = np.linalg.qr(np.random.rand(matrix_size, matrix_size))
matrix = q_matrix

# Frobenius norm
frobenius_norm = np.linalg.norm(matrix, ord="fro")

# Induced 2-norm
induced_norm_2 = np.linalg.norm(matrix, ord=2)

# Compute the induced norm via eigenvalues of A^T A
gram_matrix = matrix.T @ matrix
expected_induced_norm = np.sqrt(np.max(np.linalg.eigvals(gram_matrix).real))
assert np.allclose(induced_norm_2, expected_induced_norm)
print(f"Equal? {np.allclose(induced_norm_2, expected_induced_norm)}\n")

# Schatten p-norm (here p = 1)
p = 1
singular_values = np.linalg.svd(matrix, compute_uv=False)
schatten_p_norm = np.sum(singular_values ** p) ** (1 / p)

print(
    f"Frobenius norm: {frobenius_norm} \n"
    f"induced norm: {induced_norm_2} \n"
    f"Schatten p-norm: {schatten_p_norm}"
)
```
