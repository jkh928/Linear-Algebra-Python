## Dot Product Geometry

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('default')
```

#### The geometric equation involves the product of the length of each vector multiplied by the cosine of the angle between the two vectors.
#### $\alpha$ = ∥v∥ ∥w∥ cos(θ)

#### Features of the dot product geometry


```python
# two vectors
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])

# compute the angle (radians) between the two vectors
ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

# draw them
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b', label='Vector 1')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r', label='Vector 2')

ax.set_xlim([-6,6])
ax.set_ylim([-6,6])
ax.set_zlim([-6,6])

ax.set_box_aspect([1,1,1])

plt.title('Angle between vectors: %s rad' % ang)
plt.legend()
plt.show()
```

```python
## equivalence of algebraic and geomtric dot product formulas

# algebraic
dp_a = np.dot(v1, v2)

# geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

# print dot product
print(dp_a)
print(dp_g)
```

#### Test to see if vectors are orthogonal

```python
a = np.array([16,-2, 4])
b = np.array([.5, 2, -1])

dp_ab = np.dot(a, b)
print(dp_ab)
```
