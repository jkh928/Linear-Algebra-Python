## Dot Product Sign & Scalar Multiplication Challenge

```python
import numpy as np
```

#### The challenge is to test whether the dot product sign is invariant to scalar mulitplication

#### 1. Generate two vectors (R3)
#### 2. Generate two scalars
#### 3. Compute the dot product between vectors
#### 4. Compute the dot product between the scaled vectors

```python
# generate two vectors (R3)
v1 = np.array([-3, 4, 5])
v2 = np.array([3, 6, -3])

# generate two scalars
s1 = 2
s2 = 3

# compute dot product between vectors
print('Original dot product:' + str(np.dot(v1, v2)))

# compute dot product between scaled vectors
print('Scaled dot product:' + str(np.dot(s1*v1, s2*v2)))
```

##### The sign of the dot product is not invariant to scalar multiplication, it does have an affect on the sign of the dot product.

```python
v3 = np.array([1, 2, 3])
s3 = -2

# compute dot product between vectors
print('Original dot product:' + str(np.dot(v3, v2)))

# compute dot product between scaled vectors with a negative sign
print('Negative scaled dot product:' + str(np.dot(s3*v3, s2*v2)))
```
