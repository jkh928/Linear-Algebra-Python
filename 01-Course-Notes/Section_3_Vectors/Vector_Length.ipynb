## Calculating Vector Length (magnitude or norm)

```python
import numpy as np
```

#### Calculating the vector length is the same as taking the square root of the dot product of the vector with itself or using the pythagorean theorem
$$\|\vec{a}\| = \sqrt{\vec{a} \cdot \vec{a}}$$

```python
# Calculate vector length
v = np.array([2,3])
result = (np.sqrt(np.dot(v, v)))
print(result)
```

```python
# Another method
v1 = np.linalg.norm(v)
print(v1)
```

##### The answers are the same.
