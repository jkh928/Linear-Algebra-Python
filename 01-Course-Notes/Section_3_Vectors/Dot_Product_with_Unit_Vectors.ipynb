## Code Challenge: Dot Products with Unit Vectors
**Objective:** Determine the effect of vector magnitude on the dot product by comparing the dot product of two random vectors in $R^4$ with the dot product of their normalized unit vectors.

**Key Insight:** When we normalize vectors to a length of $1$ ($\hat{v}$), their dot product no longer scales with their original size. The dot product of two unit vectors is bounded between $-1$ and $1$, representing the cosine of the angle between them (Cosine Similarity).

```python
import numpy as np

# 1. Create two random-integer vectors (R4). Scaled by 20 to allow for more variability in the distribution of random integers.
n = 4
v1 = np.round( 20*np.random.rand(n) )
v2 = np.round( 20*np.random.rand(n) )

# 2. Compute the lengths of the individual vectors, and the magnitude of their dot product
v1m = np.sqrt(np.dot(v1,v1))
v2m = np.sqrt(np.dot(v2,v2))
dpm = np.abs(np.dot(v1,v2))
print(v1m, v2m, dpm)
print(' ')

# 3. "Normalize" the vectors (convert them to unit vectors)
v1_hat = v1/np.linalg.norm(v1)
v2_hat = v2/np.linalg.norm(v2)
print(v1_hat, v2_hat)
print(' ')

# 4. Compute the magnitude of that new dot product
dpm = np.abs(np.dot(v1_hat,v2_hat))
print(dpm)
```
