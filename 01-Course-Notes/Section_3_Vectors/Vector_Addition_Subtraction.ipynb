```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

## Vector Addition/Subtraction

 This section provides an understanding of algebraic/geometric interpretations of vector addition/subtraction

#### Vector Addition - Algebraic

Vectors being added must have the same number of elements, as each element is added with the corresponding element in the opposing vector

[1, 2] + [-2, -1] = [-1, 1]

#### Vector Addition - Geometric

The tail of vector two is added to the head of vector one and then a vector is drawn from the tail of vector one to the head of vector two.



#### Vector Subtraction - Algebraic

#### Vector Subtraction - Geometric

Vector subtraction has two different interpretations geometrically. First, as "negative addition" by multiplying vector two by minus one and then adding to vector one. Draw a vector from the tail of vector one to the head of negative vector two.



The other interpretation is a "head to head," where a vector is drawn from the head of vector one to the head of vector two




```python
# two vectors in R2
v1 = np.array([ 3, -1 ])
v2 = np.array([ 2,  4 ])

# add the two vectors
v3 = v1 + v2


# plot them
plt.quiver(0, 0, v1[0], v1[1], color='b', angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(v1[0], v1[1], v2[0], v2[1], color='r', angles='xy', scale_units='xy', scale=1, label='v2')
plt.quiver(0, 0, v3[0], v3[1], color='k', angles='xy', scale_units='xy', scale=1, label='v1+v2')

plt.legend()
plt.title('Vector Addition')
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid(True)
plt.show()
```

```python
# subtract vector two from vector one
v4 = v1 - v2


# plot them
plt.figure(figsize=(6,6))

# quiver(origin_x, origin_y, change_x, change_y)
plt.quiver(0, 0, v1[0], v1[1], color='b', angles='xy', scale_units='xy', scale=1, label='v1')
plt.quiver(v1[0], v1[1], -v2[0], -v2[1], color='r', angles='xy', scale_units='xy', scale=1, label='-v2')
plt.quiver(0, 0, v4[0], v4[1], color='k', angles='xy', scale_units='xy', scale=1, label='v1-v2')

plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid(True)
plt.legend()
plt.title('Vector Subtraction')
plt.show()
```
