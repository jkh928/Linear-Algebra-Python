## Subspaces
#### 1. Definition: A subspace of $\mathbb{R}^n$ is any set $H$ in $\mathbb{R}^n$ that possesses three specific properties:The zero vector is in $H$. For each $\mathbf{u}$ and $\mathbf{v}$ in $H$, the sum $\mathbf{u} + \mathbf{v}$ is in $H$ (Closure under addition). For each $\mathbf{u}$ in $H$ and each scalar $c$, the vector $c\mathbf{u}$ is in $H$ (Closure under scalar multiplication).
#### 2. The "Flatness" ConceptSubspaces are essentially "linear" or "flat" subsets that must pass through the origin. If a set of vectors forms a line or a plane but does not include the origin, it fails the first property and is therefore not a subspace.

```python
def verify_subspace_properties(vectors, scalar_list=[-1, 0, 2, 0.5]):
    """
    Tests a collection of vectors for basic subspace closure properties.
    """
    vecs = [np.array(v) for v in vectors]

    # 1. Check for Zero Vector (approximate for floats)
    has_zero = any(np.allclose(v, 0) for v in vecs)
    print(f"Contains Zero Vector: {has_zero}")

    # 2. Check Additive Closure (Sample test)
    # Test if u + v is in the sample set
    u, v = vecs[0], vecs[1]
    w = u + v
    # Check if w is 'close' to any vector in the set
    is_additive_closed = any(np.allclose(w, x) for x in vecs)
    print(f"Additive Closure (Sample): {is_additive_closed}")

    # 3. Check Scalar Closure (Sample test)
    # Test if c * u is in the sample set
    c = scalar_list[0]
    scaled_v = c * u
    is_scalar_closed = any(np.allclose(scaled_v, x) for x in vecs)
    print(f"Scalar Closure (Sample): {is_scalar_closed}")

# --- Example 1: A valid Subspace (Line through origin: y = x) ---
subspace_sample = [[0,0], [1,1], [2,2], [-1,-1], [0.5, 0.5]]
print("--- Testing Line y=x (Subspace) ---")
verify_subspace_properties(subspace_sample)
```

## Subspaces vs. Subsets
#### 1. Subset vs. SubspaceSubset: Any collection of vectors from $\mathbb{R}^n$.Subspace: A specific type of subset that satisfies the three rules of closure and contains the origin.
#### 2. Geometric Examples in $\mathbb{R}^3$The Origin: A zero-dimensional subspace.Lines through the Origin: One-dimensional subspaces.Planes through the Origin: Two-dimensional subspaces. The Entire Space: $\mathbb{R}^3$ is a subspace of itself.
#### 3. Common Non-Subspaces (Subsets only)A line that misses the origin (Fails Rule 1). A "chunk" or segment of a line (Fails Rule 3: scaling a vector could push it outside the segment). The first quadrant of $\mathbb{R}^2$ (Fails Rule 3: multiplying by a negative scalar pushes the vector into the third quadrant).

```python
# --- Example 2: A Subset (The first quadrant: x>0, y>0) ---
subset_sample = [[1,1], [2,2], [5,10]] # No zero, no negatives
print("\n--- Testing First Quadrant (Subset Only) ---")
verify_subspace_properties(subset_sample)
```
