## Unit Vectors ($\hat{v}$)
A unit vector is a vector that has a magnitude (length) of exactly $1$. It is used purely to indicate **direction**, stripping away the "weight" or length of the original data.

To create a unit vector, you simply divide the original vector by its own magnitude. Mathematically, this is written as:
$$\hat{v} = \frac{v}{||v||}$$

*The code below calculates the unit vector $\hat{v}$ and plots it against the original vector $v$ to visually prove that they point in the exact same direction, but $\hat{v}$ stops at a length of $1$.*

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')

# 1. The Math (from our previous step)
v = np.array([3, 4])
v_hat = v / np.linalg.norm(v)

# 2. Set up the canvas
plt.figure(figsize=(6, 6))

# 3. Plot the original vector (v) as a solid blue line
# Syntax: plt.plot([x_start, x_end], [y_start, y_end])
plt.plot([0, v[0]], [0, v[1]], 'b-', label='Original Vector (v)', linewidth=2)

# 4. Plot the unit vector (v_hat) as a dashed red line
plt.plot([0, v_hat[0]], [0, v_hat[1]], 'r--', label='Unit Vector (v-hat)', linewidth=4)

# 5. Format the graph for perfect geometry
plt.axis('square')         # Prevents distortion so a length of 1 actually looks like 1
plt.axis([-1, 5, -1, 5])   # Sets the boundaries of the graph
plt.grid(True)             # Turns on the background grid
plt.legend()               # Adds the legend
plt.title("Original Vector vs. Unit Vector")

# 6. Render the plot
plt.show()
```
