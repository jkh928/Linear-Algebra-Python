## Foundations : 2D and 3D Vector Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

## Algebraic Vectors

Definitions:

Vector - An ordered list of numbers

Element - Each number inside the vector

Dimensionality - The number of elements in a vector

## Geometric vectors

Definition:

A vector is a straight line with some length and some direction, specified by the numbers given in a vector



## 2D Foundations:
 Visualizing vectors in $\mathbb{R}^2$ (X, Y).

```python
# To create a vector for simple operations in Python, use lists.
#2D vector
v2 = [3, -2]

# 3D vector
v3 = [4, -3, 2]

#To transpose a vector in a list (switch the rows and columns)
# row to column (or vice versa):
v3t = np.transpose(v3)

#Execute the plot command using the [x,y] coordinates, providing
# a start and end point for each. In the example, x-coordinates
# start at x=0 and end at y=v2[0] (vector 2, first element), and
# y-coordinates start at y=0 and end at y=v2[1] (vector 2, second
# element)

# plot the 2D vector
plt.plot([0,v2[0]],[0,v2[1]], label ='Vector v2')

# set the x and y axes (k = black and -- = dashed lines for clarity)
plt.plot([-4, 4],[0, 0],'k--')
plt.plot([0, 0],[-4, 4],'k--')
# show grid
plt.grid(True)
# set the scale
plt.axis((-4, 4, -4, 4))
# force the 1:1 aspect ratio
plt.gca().set_aspect('equal', adjustable='box')
# render the image
plt.show()
```

## 3D Visualizations:
Visualizing vectors in $\mathbb{R}^3$ (X, Y, Z).

```python
# plot the 3D vector. plt.figure() creates the window and
# plt.figaspect(1) sets the window to a perfect square at ratio of 1
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d') # another way to accomplish the
# code below. 'projection = '3d'' initializes Axes3D toolkit to change
# coordinate system from (x,y) to (x,y,z).
# ax = plt.axes(projection = '3d')
#plot the x,y,z coordinates and specify linewidth (thicker)
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]],linewidth=3, label ='Vector v3')
ax.plot([0, 0],[0, 0],[-4, 4],'k--', alpha=0.5) #Z
ax.plot([0, 0],[-4, 4],[0, 0],'k--', alpha=0.5) #Y
ax.plot([-4, 4],[0, 0],[0, 0],'k--', alpha=0.5) #X

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Vector Representation')

fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax.set_box_aspect(None, zoom=0.85) # Zoom out the box slightly

plt.show()
```
