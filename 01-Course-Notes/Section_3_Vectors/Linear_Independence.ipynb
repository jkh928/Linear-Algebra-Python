# Linear Algebra: Linear Independence

### **1. Core Definition**
A set of vectors $\{v_1, v_2, \dots, v_n\}$ is **linearly independent** if the only solution to the vector equation:

$$c_1v_1 + c_2v_2 + \dots + c_nv_n = \mathbf{0}$$

is the **trivial solution**:

$$c_1 = c_2 = \dots = c_n = 0$$

> **Note:** If there is any other way to get to the zero vector (where at least one $c$ is not zero), the set is **Linearly Dependent**.

---

### **2. The Geometric Perspective**
* **Independent:** Each vector points in a "new" direction not covered by the others. They don't lie on the same line (if 2 vectors) or the same plane (if 3 vectors).
* **Dependent:** At least one vector is "redundant." It can be created by stretching and adding the other vectors in the set.



---

### **3. Key Observation Rules**
1. **The Zero Vector:** Any set containing the zero vector ($\mathbf{0}$) is **automatically linearly dependent** because you can multiply $\mathbf{0}$ by any number to stay at zero.
2. **The Proportion Rule:** Two vectors are dependent if one is simply a scalar multiple of the other (e.g., $[1, 2]$ and $[2, 4]$).
3. **The "Too Many Vectors" Rule:** If you have more vectors than you have dimensions (e.g., 3 vectors in $\mathbb{R}^2$), the set **must** be linearly dependent.

---

### **4. Engineering Relevance**
* **Redundancy:** In data sets, linear dependence means you have columns of data that provide no new information—they are just combinations of other columns.
* **Dimensionality:** Linear independence defines the "basis" of a space; it tells you the minimum number of vectors needed to reach every point in that coordinate system.
