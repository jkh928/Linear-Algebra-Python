## Scalar Multiplication

This section covers how to multiply a vector by a scalar.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

#### Algebraic

Multiply each element of the vector by the scalar

#### Geometric

Stretch or shrink the vector by the scalar



Scalar > 1 The vector will be longer

0 < Scalar < 1 The vector will be smaller than the original vector

Scalar < 0 The vector points in the opposite direction.
