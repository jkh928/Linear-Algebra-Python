## The Dot Product - Vectors

This section covers how to find the dot product of vectors.



```python
import numpy as np
import matplotlib.pyplot as plt
```

a<sup>T</sup>b = a transpose b (where a and b are column vectors)

The dot product is the sum of the product of each corresponding element in each vector

```python
## there are many ways to compute the dot product

v1 = np.array([1, 2, 3, 4, 5])
v2 = np.array([0, -4, -3, 6, 5])

# method 1
dp1 = sum( np.multiply(v1, v2) )

# method 2 (preferred method)
dp2 = np.dot( v1, v2 )

# method 3
dp3 = np.matmul( v1, v2 )

# method 4
dp4 = 0 # initialize

# loop over elements
for i in range(len(v1)):
    dp4 = dp4 + v1[i] * v2[i]


print(dp1, dp2, dp3, dp4)
```
