## Cauchy-Schwarz Inequality Challenge

```python
import numpy as np

# create three vectors, one is dependent on another
a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(1) * a

# compute dot products
aTb = np.dot(a,b)
aTc = np.dot(a,c)

# demonstrate the (in)equalities
print(f' {np.abs(aTb):.4f}, {np.linalg.norm(a)*np.linalg.norm(b):.4f}')
print(f' {np.abs(aTc):.4f}, {np.linalg.norm(a)*np.linalg.norm(c):.4f}')
```

##### A and C form a linearly dependent pair, which means the cosine of the angle between them are either minus one or plus one.

```python
## determine which vector pairs meet at an obtuse angle

# create vectors
d = np.array([1, -2])
e = np.array([2, 3])
f = np.array([0, 2])

# calculate dot products for all pairs
dp_de = np.dot(d,e)
dp_df = np.dot(d,f)
dp_ef = np.dot(e,f)

# show results
print(f"d * e = {dp_de} --> {'Obtuse' if dp_de < 0 else 'Acute/Right'}")
print(f"d * f = {dp_df} --> {'Obtuse' if dp_df < 0 else 'Acute/Right'}")
print(f"e * f = {dp_ef} --> {'Obtuse' if dp_ef< 0 else 'Acute/Right'}")

```
