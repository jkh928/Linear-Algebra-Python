## Theory: Dimensions and Fields
Before building complex matrices, we must define the rules of the "sandbox" we are playing in.

### 1. Fields ($\mathbb{F}$)
A field defines the set of numbers we are allowed to use for our scalars and vector elements.
* **$\mathbb{R}$ (Real Numbers):** The standard field in machine learning (integers, fractions, irrational numbers).
* **$\mathbb{C}$ (Complex Numbers):** Numbers that include an imaginary component ($i$), which we handled using the Hermitian transpose.

### 2. Dimensions ($n$)
The dimension of a vector simply refers to the number of elements it contains. This dictates the geometric "ambient space" the vector lives in.
* A vector with 2 elements lives in $\mathbb{R}^2$ (a 2D plane).
* A vector with 3 elements lives in $\mathbb{R}^3$ (a 3D volume).
* A vector with $n$ elements lives in $\mathbb{R}^n$ (an $n$-dimensional space).

*Note: In Python, we find the dimension of a 1D NumPy array using `len(v)` or `v.shape[0]`.*
