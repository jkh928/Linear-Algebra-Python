## Is Dot Product Commutative? Challenge

```python
import numpy as np
```

#### 1. Generate two 100-element random row vectors, compute dot product a with b, b with a

```python
# Part 1: 100-element random vectors (Floats)
# We use 1D arrays which numpy treats as vectors
a = np.random.randn(100)
b = np.random.randn(100)

dp_ab = np.dot(a,b)
dp_ba = np.dot(b,a)
print(dp_ab, dp_ba, dp_ab - dp_ba)
print(' ')
```

 #### Generate two 2-element integer row vectors, repeat.

```python
# Part 2: 2-element random vectors (Integers)
c = np.random.randint(0, 10, 2)
d = np.random.randint(0, 10, 2)
print(c)
print(d)
print(np.dot(c,d), np.dot(d,c))
```

##### Based on this challenge, the dot product is indeed commutative.
